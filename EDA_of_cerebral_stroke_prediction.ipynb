```python
# Fill missing values in 'smoking_status' with 'Unknown'
missing_smoking = cleaned_missing_value_handled_id_dropped_smoking_status_dropna_stroke_df['smoking_status'].isna().sum()
print(f"Missing values in 'smoking_status': {missing_smoking} out of {len(cleaned_missing_value_handled_id_dropped_smoking_status_dropna_stroke_df)} records")
cleaned_missing_value_handled_id_dropped_smoking_status_dropna_stroke_df['smoking_status'] = cleaned_missing_value_handled_id_dropped_smoking_status_dropna_stroke_df['smoking_status'].fillna('Unknown')

# Compute average glucose level by work type and smoking status
glucose_by_work_smoking = cleaned_missing_value_handled_id_dropped_smoking_status_dropna_stroke_df.groupby(['work_type', 'smoking_status'])['avg_glucose_level'].mean().reset_index()
pivot_table = glucose_by_work_smoking.pivot(index='work_type', columns='smoking_status', values='avg_glucose_level')

print("\nMean Glucose Level Grouped by Work Type and Smoking Habit:")
print(pivot_table)

# Visualize results
plt.figure(figsize=(14, 8))
ax = sns.barplot(x='work_type', y='avg_glucose_level', hue='smoking_status', data=glucose_by_work_smoking, palette='viridis')

plt.title('Avg Glucose Levels by Work Type & Smoking Status', fontsize=16)
plt.xlabel('Work Type', fontsize=14)
plt.ylabel('Avg Glucose Level (mg/dL)', fontsize=14)
plt.legend(title='Smoking Status', fontsize=12)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

# Annotate bars with exact values
for bar in ax.patches:
    height = bar.get_height()
    if not np.isnan(height):
        ax.text(bar.get_x() + bar.get_width() / 2, height + 0.3, f'{height:.1f}', ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('images/glucose_levels_by_work_and_smoking_status.png')
plt.show()
```
