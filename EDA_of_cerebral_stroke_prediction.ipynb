# Step 6: Correlation Analysis

```python
# Calculate correlations only on numerical features
numerical_features = clean_data.select_dtypes(include=np.number).columns
correlation_matrix = clean_data[numerical_features].corr()

# Display the correlation matrix
print(correlation_matrix)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.savefig('images/correlation_matrix.png')
plt.show()
```

# Pie chart of positive correlations

```python
correlation_with_stroke = clean_data.corr(numeric_only=True)['stroke'].sort_values(ascending=False)
positive_corr = correlation_with_stroke[correlation_with_stroke >= 0]
positive_corr.plot(kind='pie', figsize=(8, 8))
plt.title('Positive Correlations with Stroke')
plt.savefig('images/positive_correlations_stroke.png')
plt.show()
```

# Step 7: Visualizing Distributions and Relationships

# Distribution of target variable 'stroke

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='stroke', data=clean_data, palette='Set2')
plt.title('Distribution of Stroke')
plt.xlabel('Stroke')
plt.ylabel('Count')
plt.savefig('images/stroke_distribution.png')
plt.show()
```

# Stroke occurrence by age

```python
# Stroke occurrence by age
age_stroke = clean_data.groupby('age')['stroke'].mean().reset_index()
fig = px.line(age_stroke, x='age', y='stroke',
              title='Stroke Occurrence by Age',
              labels={'stroke': 'Probability of Stroke', 'age': 'Age'},
              markers=True)
fig.update_traces(line=dict(width=2))  # Thicker line for visibility
fig.write_image('images/stroke_occurrence_by_age.png')
fig.show()
```

# Stroke probability by work type status



```python
# Stroke probability by work type
worktype_stroke_probabilities = clean_data.groupby("work_type")["stroke"].mean().reset_index()
print("\nStroke Probabilities by Work Type:")
print(worktype_stroke_probabilities)
fig = px.bar(worktype_stroke_probabilities, x='work_type', y='stroke',
             title='Work Type vs Stroke Probability',
             labels={'stroke': 'Stroke Probability', 'work_type': 'Work Type'},
             color='work_type')
fig.write_image('images/work_type_stroke_probability.png')
fig.show()
```

### Stroke probability by smoking status

```python
# Stroke probability by smoking status
smoking_stroke_probabilities = clean_data.groupby("smoking_status")["stroke"].mean().reset_index()
print("\nStroke Probabilities by Smoking Status:")
print(smoking_stroke_probabilities)
fig = px.pie(smoking_stroke_probabilities, values='stroke',
             names='smoking_status',
             title='Stroke Probability by Smoking Status',
             color_discrete_sequence=px.colors.sequential.OrRd_r)
fig.write_image('images/smoking_status_stroke_probability.png')
fig.show()
```

# Stroke risk by age brackets




```python
# Stroke risk by age brackets
plt.figure(figsize=(15, 10))  # Increase figure size for better spacing
age_bins = [0, 18, 30, 45, 60, 75, 90]
clean_data['age_group'] = pd.cut(clean_data['age'], bins=age_bins)
age_stroke = clean_data.groupby('age_group')['stroke'].mean().reset_index()
plt.subplot(2, 2, 1)
sns.barplot(x='age_group', y='stroke', data=age_stroke, palette='viridis')
plt.title('Stroke Probability by Age Group')
plt.xticks(rotation=45)
plt.xlabel('Age Group')
plt.ylabel('Stroke Probability')
plt.savefig('images/stroke_probability_by_age_group.png')
plt.show()
```

# Hypertension & heart disease interaction




```python
# Hypertension & heart disease interaction
conditions = clean_data.groupby(['hypertension', 'heart_disease'])['stroke'].mean().reset_index()
conditions['condition'] = conditions.apply(lambda x:
    "Both+" if x['hypertension'] and x['heart_disease'] else
    "Hypertension+" if x['hypertension'] else
    "Heart Disease+" if x['heart_disease'] else
    "None", axis=1)
print("Stroke Probabilities by Medical Conditions:")
print(conditions)  # Verify data
plt.subplot(2, 2, 2)
sns.barplot(x='condition', y='stroke', data=conditions, palette='Set2')  # Switch to Set2 for distinct colors
plt.title('Stroke Risk by Medical Conditions')
plt.xlabel('Condition')
plt.ylabel('Stroke Probability')
plt.xticks(rotation=45, ha='right')  # Rotate and align labels to prevent overlap
plt.savefig('images/stroke_risk_by_medical_conditions.png')
plt.show()
```

# Glucose level impact

```python
# Glucose level impact
glucose_bins = [0, 140, 200, 300]
clean_data['glucose_category'] = pd.cut(clean_data['avg_glucose_level'], bins=glucose_bins,
                                        labels=['Normal', 'Prediabetes', 'Diabetes'])
glucose_stroke = clean_data.groupby('glucose_category')['stroke'].mean().reset_index()
plt.subplot(2, 2, 3)
sns.barplot(x='glucose_category', y='stroke', data=glucose_stroke, palette='Blues')
plt.title('Stroke Risk by Glucose Levels')
plt.xlabel('Glucose Category')
plt.ylabel('Stroke Probability')
plt.xticks(rotation=45, ha='right')
plt.savefig('images/glucose_level.png')
plt.show()
```

# Stroke probability for hypertension patients





```python
# Stroke probability for hypertension patients
hypertension_df = clean_data[clean_data['hypertension'] == 1]
hypertension_stroke_prob = hypertension_df.groupby(['gender', 'age'])['stroke'].mean().reset_index()
print("\nStroke Probability for Individuals with Hypertension:")
print(hypertension_stroke_prob)
fig = px.density_heatmap(hypertension_stroke_prob, x='age', y='gender', z='stroke',
                         title='Stroke Probability by Age and Gender (With Hypertension)',
                         labels={'stroke': 'Stroke Probability'},
                         color_continuous_scale='Reds')
fig.write_image('images/stroke_probability_by_age_and_gender_with_hypertension.png')
fig.show()
```

# Age group analysis for stroke occurrences

```python
# Age group analysis for stroke occurrences
age_bins = [0, 18, 35, 50, 65, 100]
age_labels = ['<18', '18-35', '35-50', '50-65', '65+']
clean_data['age_group'] = pd.cut(clean_data['age'], bins=age_bins, labels=age_labels, right=False)
stroke_data = clean_data[clean_data['stroke'] == 1]
plt.figure(figsize=(12, 7))
ax = sns.countplot(data=stroke_data, x='age_group', hue='gender', palette='Set2', order=age_labels)
plt.title('Distribution of Stroke Occurrences across Age Groups and Genders', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Stroke Count', fontsize=12)
plt.legend(title='Gender', fontsize=10)
for p in ax.patches:
    height = p.get_height()
    if not np.isnan(height) and height > 0:
        ax.text(p.get_x() + p.get_width()/2., height + 0.1,
                int(height),
                ha="center", fontsize=10, fontweight='bold')
plt.tight_layout()
plt.savefig('images/age_gender_stroke_occurrences.png')
plt.show()
```

```python
# Handle missing values in smoking_status (moved from preprocessing if not already done)
print("Missing smoking_status values: " + str(clean_data['smoking_status'].isna().sum()) + " out of " + str(len(clean_data)) + " rows")
clean_data['smoking_status'] = clean_data['smoking_status'].fillna('Unknown')
glucose_by_work_smoking = clean_data.groupby(['work_type', 'smoking_status'])['avg_glucose_level'].mean().reset_index()
pivot_table = glucose_by_work_smoking.pivot(index='work_type', columns='smoking_status', values='avg_glucose_level')
print("\nAverage Glucose Levels by Work Type and Smoking Status:")
print(pivot_table)
plt.figure(figsize=(14, 8))
ax = sns.barplot(x='work_type', y='avg_glucose_level', hue='smoking_status', data=glucose_by_work_smoking, palette='viridis')
plt.title('Impact of Work Type and Smoking Status on Average Glucose Levels', fontsize=16)
plt.xlabel('Work Type', fontsize=14)
plt.ylabel('Average Glucose Level (mg/dL)', fontsize=14)
plt.legend(title='Smoking Status', fontsize=12)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
for p in ax.patches:
    height = p.get_height()
    if not np.isnan(height):
        ax.text(p.get_x() + p.get_width()/2., height + 0.3, f'{height:.1f}', ha="center", fontsize=10)
plt.tight_layout()
plt.savefig('images/glucose_by_work_smoking.png')
plt.show()
```

###Stroke Probability by Smoking Status, BMI, and Glucose Levels

```python
# Define BMI categories (consistent with #9)
bmi_bins = [0, 18.5, 25, 30, 100]
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
clean_data['bmi_bracket'] = pd.cut(clean_data['bmi'], bins=bmi_bins, labels=bmi_labels, right=False)

# Define glucose level categories (aligned with #8 for consistency)
glucose_bins = [0, 140, 200, 300]  # Match #8
glucose_labels = ['Normal', 'Prediabetes', 'Diabetes']
clean_data['glucose_bracket'] = pd.cut(clean_data['avg_glucose_level'], bins=glucose_bins, labels=glucose_labels)

# Ensure smoking_status has no missing values (from #14)
clean_data['smoking_status'] = clean_data['smoking_status'].fillna('Unknown')

# Grouping
stroke_by_smoking_bmi_glucose = clean_data.groupby(['smoking_status', 'bmi_bracket', 'glucose_bracket'])['stroke'].mean().reset_index()
filtered_df = stroke_by_smoking_bmi_glucose[stroke_by_smoking_bmi_glucose['glucose_bracket'] == 'Diabetes']

# Plot
plt.figure(figsize=(16, 8))
ax = sns.barplot(x='smoking_status', y='stroke', hue='bmi_bracket', data=filtered_df, palette='coolwarm')
plt.title('Stroke Probability by Smoking Status and BMI (Diabetes Only)')
plt.xlabel('Smoking Status')
plt.ylabel('Stroke Probability')
plt.xticks(rotation=45)
for p in ax.patches:
    height = p.get_height()
    if not np.isnan(height):
        ax.annotate(f'{height:.2f}',
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                    textcoords='offset points')
plt.legend(title='BMI Category')
plt.tight_layout()
plt.savefig('images/stroke_probability_by_smoking_bmi_glucose.png')
plt.show()
```
